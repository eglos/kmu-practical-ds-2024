# Docker 설치 및 실행

URL : https://www.docker.com/products/docker-desktop

# Docker Image 다운로드

- 맥은 terminal, 윈도우는 cmd 창을 열고 다음 명령어를 실행합니다.
- docker 명령어를 실행할때 위치한 경로는 위치한 경로는 각자의 폴더 하위에 workspace 라는 폴더를 만들어 사용하기를 권장합니다.

```bash
$ docker pull jupyter/datascience-notebook
```

# Docker 컨테이너 실행

* Mac 유저
```bash
$ docker run -d -p 10000:8888 --name jupyter -v ${PWD}:/home/jovyan/work quay.io/jupyter/datascience-notebook:2024-03-14
```

* Windows 유저
```bash
$ docker run -d -p 10000:8888 --name jupyter -v "%cd%":/home/jovyan/work quay.io/jupyter/datascience-notebook:2024-03-14
```

# Jupyter 실행

* Chrome 브라우저를 열고, 주소창에 https://localhost:10000 을 입력
* 출력된 token을 확인하여, New password 생성
* New password 입력 후, Jupyter 접속

# Docker 컨테이너 중지
* Docker 컨테이너를 중지하고 싶을 때는 아래 명령어를 실행한다.
```bash
$ docker stop jupyter
```

# Docker 컨테이너 재실행
* Docker 컨테이너를 재실행하면 다시 크롬으로 주피터가 접속가능하다.
```bash
$ docker start jupyter
```

# Docker 컨테이너 완전 삭제
* Docker 컨테이너를 삭제하면 추후 다시 이용시, docker run 명령어로 컨테이너를 다시 생성해야한다.
```bash
$ docker rm jupyter
```

# Docker 이미지 삭제
* 이미지를 삭제하면, 다시 사용하고 싶은 경우, 이미지를 재생성하거나 다운로드(Push)해야 한다.
```bash
$ docker rmi quay.io/jupyter/datascience-notebook:2024-03-14
```

# 사용자가 설치한 패키지를 Docker 이미지에 영구 적용하고 싶을 때
* Docker 컨테이너를 중지하고 싶을 때는 아래 명령어를 실행한다.
```bash
$ docker commit jupyter your/image/name:your/tag/name <<-- 다른 이름의 이미지로 로컬 저장
```
